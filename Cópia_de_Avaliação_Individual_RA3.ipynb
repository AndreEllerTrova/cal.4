Em uma região desértica, pesquisadores estão estudando o crescimento de uma espécie rara de cacto. O crescimento do cacto ao longo do tempo é modelado pela função \( y = x^5 (2 - x) \), onde \( y \) representa a altura do cacto em metros e \( x \) representa o tempo em anos desde o início do estudo. Os pesquisadores querem determinar a quantidade total de biomassa acumulada pelo cacto ao longo dos primeiros 2 anos. Sabemos, de estudos anteriores, que a biomassa acumulada é diretamente proporcional (1:1) à área sob a curva de crescimento do cacto ao longo do tempo. Determine a quantidade total de biomassa acumulada nestes dois anos.

### Solução

Para determinar a quantidade total de biomassa acumulada pelo cacto ao longo dos primeiros 2 anos, precisamos calcular a área sob a curva da função \( y = x^5 (2 - x) \) no intervalo de \( x = 0 \) a \( x = 2 \). Isso corresponde à integral definida da função \( y = x^5 (2 - x) \) de 0 a 2.

1. A função a ser integrada é:

$$
y = x^5 (2 - x)
$$

2. A integral definida de \( y \) de 0 a 2 é:

$$
\text{Biomassa acumulada} = \int_{0}^{2} x^5 (2 - x) \, dx
$$

3. Expanda a função para simplificar a integração:

$$
\int_{0}^{2} x^5 (2 - x) \, dx = \int_{0}^{2} (2x^5 - x^6) \, dx
$$

4. Separe em duas integrais:

$$
= \int_{0}^{2} 2x^5 \, dx - \int_{0}^{2} x^6 \, dx
$$

5. Integre cada termo separadamente:

$$
\int 2x^5 \, dx = 2 \int x^5 \, dx = 2 \left( \frac{x^6}{6} \right) = \frac{2x^6}{6} = \frac{x^6}{3}
$$

$$
\int x^6 \, dx = \frac{x^7}{7}
$$

6. Avalie as integrais de 0 a 2:

$$
\left[ \frac{x^6}{3} \right]_{0}^{2} - \left[ \frac{x^7}{7} \right]_{0}^{2}
$$

$$
= \left( \frac{2^6}{3} - \frac{0^6}{3} \right) - \left( \frac{2^7}{7} - \frac{0^7}{7} \right)
$$

$$
= \frac{64}{3} - \frac{128}{7}
$$

7. Calcule a diferença:

$$
\frac{64}{3} - \frac{128}{7} = \frac{64 \cdot 7}{3 \cdot 7} - \frac{128 \cdot 3}{7 \cdot 3}
$$

$$
= \frac{448}{21} - \frac{384}{21}
$$

$$
= \frac{448 - 384}{21}
$$

$$
= \frac{64}{21}
$$

Portanto, a quantidade total de biomassa acumulada pelo cacto nos primeiros 2 anos é:

$$
\frac{64}{21} \, \text{unidades de biomassa}
$$




# Questão 2 - Vale até 1 ponto
Usando o Python (matplotlib, sympy e numpy) trace o gráfico e calcule a área sob a função $F(x) = x^5(2 - x)$ para $0 \leq x \leq 2$. Não esqueça de anotar os pontos, hachurar a área calculada e de imprimir o valor da área encontrada.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definindo a função
def F(x):
    return x**5 * (2 - x)

# Valores de x
x = np.linspace(0, 2, 400)
y = F(x)

# Plotando a função
plt.plot(x, y, label=r'$F(x) = x^5(2 - x)$')
plt.fill_between(x, y, where=[(xi >= 0) and (xi <= 2) for xi in x], color='skyblue', alpha=0.4)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Calculando a área sob a curva usando sympy
x_sym = sp.Symbol('x')
F_sym = x_sym**5 * (2 - x_sym)
area = sp.integrate(F_sym, (x_sym, 0, 2))
print(f'Área sob a curva de 0 a 2: {area.evalf()}')

# Marcando pontos e hachurando a área
plt.annotate(f'Área = {area.evalf():.2f}', xy=(1, 1), xytext=(1.5, 4),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('Área sob a função $F(x) = x^5(2 - x)$ de 0 a 2')
plt.xlabel('$x$')
plt.ylabel('$F(x)$')
plt.show()
```

A aceleração de um objeto é dada por \(a(t) = \cos(\pi t)\), e sua velocidade no tempo \(t = 0\) é \(\frac{1}{2\pi}\). Usando apenas álgebra, encontre a distância total percorrida por este objeto nos primeiros \(1,5\) segundos deste movimento. Dica: antes de calcular qualquer coisa, verifique cuidadosamente a equação da aceleração. Não é necessário, mas talvez seja interessante, traçar um gráfico da aceleração e outro da velocidade para entender o problema. É necessário entender o movimento do objeto para saber quanto espaço ele realmente percorreu.

### Solução

Para encontrar a distância total percorrida pelo objeto, precisaremos integrar a aceleração para obter a velocidade e, em seguida, integrar a velocidade para obter a posição.

#### 1. Encontrando a Velocidade

A aceleração é a derivada da velocidade. Portanto, a velocidade \(v(t)\) pode ser encontrada integrando a aceleração \(a(t)\):

$$
a(t) = \cos(\pi t)
$$

Integre \(a(t)\) em relação a \(t\):

$$
v(t) = \int a(t) \, dt = \int \cos(\pi t) \, dt
$$

Use a substituição \(u = \pi t\):

$$
du = \pi \, dt \quad \Rightarrow \quad dt = \frac{1}{\pi} \, du
$$

Substituindo na integral:

$$
v(t) = \int \cos(u) \cdot \frac{1}{\pi} \, du = \frac{1}{\pi} \int \cos(u) \, du = \frac{1}{\pi} \sin(u) + C = \frac{1}{\pi} \sin(\pi t) + C
$$

Use a condição inicial \(v(0) = \frac{1}{2\pi}\) para encontrar a constante \(C\):

$$
v(0) = \frac{1}{\pi} \sin(0) + C = \frac{1}{2\pi} \quad \Rightarrow \quad 0 + C = \frac{1}{2\pi} \quad \Rightarrow \quad C = \frac{1}{2\pi}
$$

Portanto, a função velocidade é:

$$
v(t) = \frac{1}{\pi} \sin(\pi t) + \frac{1}{2\pi}
$$

#### 2. Encontrando a Posição

A velocidade é a derivada da posição. Portanto, a posição \(s(t)\) pode ser encontrada integrando a velocidade \(v(t)\):

$$
s(t) = \int v(t) \, dt = \int \left( \frac{1}{\pi} \sin(\pi t) + \frac{1}{2\pi} \right) \, dt
$$

Divida a integral em duas partes:

$$
s(t) = \frac{1}{\pi} \int \sin(\pi t) \, dt + \frac{1}{2\pi} \int 1 \, dt
$$

Para a primeira integral, use a substituição \(u = \pi t\):

$$
\int \sin(\pi t) \, dt = -\frac{1}{\pi} \cos(\pi t) + C_1
$$

Para a segunda integral:

$$
\int 1 \, dt = t + C_2
$$

Combine os resultados:

$$
s(t) = \frac{1}{\pi} \left( -\frac{1}{\pi} \cos(\pi t) \right) + \frac{1}{2\pi} t + C = -\frac{1}{\pi^2} \cos(\pi t) + \frac{1}{2\pi} t + C
$$

Use a condição inicial \(s(0) = 0\) para encontrar a constante \(C\):

$$
s(0) = -\frac{1}{\pi^2} \cos(0) + \frac{1}{2\pi} \cdot 0 + C = 0 \quad \Rightarrow \quad -\frac{1}{\pi^2} + C = 0 \quad \Rightarrow \quad C = \frac{1}{\pi^2}
$$

Portanto, a função posição é:

$$
s(t) = -\frac{1}{\pi^2} \cos(\pi t) + \frac{1}{2\pi} t + \frac{1}{\pi^2}
$$

#### 3. Calculando a Distância Total Percorrida

Para calcular a distância total percorrida, precisamos avaliar a variação da posição \(s(t)\) de \(t = 0\) a \(t = 1.5\):

$$
\Delta s = s(1.5) - s(0)
$$

Substituindo \(t = 1.5\):

$$
s(1.5) = -\frac{1}{\pi^2} \cos(\pi \cdot 1.5) + \frac{1}{2\pi} \cdot 1.5 + \frac{1}{\pi^2}
$$

$$
\cos(\pi \cdot 1.5) = \cos\left(\frac{3\pi}{2}\right) = 0
$$

Portanto,

$$
s(1.5) = 0 + \frac{1.5}{2\pi} + \frac{1}{\pi^2} = \frac{1.5}{2\pi} + \frac{1}{\pi^2}
$$

$$
s(1.5) = \frac{1.5}{2\pi} + \frac{1}{\pi^2}
$$

Finalmente, a variação da posição de \(t = 0\) a \(t = 1.5\):

$$
\Delta s = s(1.5) - s(0) = \left(\frac{1.5}{2\pi} + \frac{1}{\pi^2}\right) - 0
$$

$$
\Delta s = \frac{1.5}{2\pi} + \frac{1}{\pi^2}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definindo as funções
t = sp.Symbol('t')
a = sp.cos(sp.pi * t)
v = sp.integrate(a, t) + 1/(2*sp.pi)
s = sp.integrate(v, t) + 1/(sp.pi**2)

# Funções lambda para numpy
a_func = sp.lambdify(t, a, 'numpy')
v_func = sp.lambdify(t, v, 'numpy')
s_func = sp.lambdify(t, s, 'numpy')

# Valores de t
t_vals = np.linspace(0, 1.5, 400)

# Calculando os valores de aceleração, velocidade e posição
a_vals = a_func(t_vals)
v_vals = v_func(t_vals)
s_vals = s_func(t_vals)

# Plotando a aceleração
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(t_vals, a_vals, label=r'$a(t) = \cos(\pi t)$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Aceleração')
plt.xlabel('$t$')
plt.ylabel('$a(t)$')

# Plotando a velocidade
plt.subplot(1, 2, 2)
plt.plot(t_vals, v_vals, label=r'$v(t)$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Velocidade')
plt.xlabel('$t$')
plt.ylabel('$v(t)$')

plt.tight_layout()
plt.show()

# Calculando a distância total percorrida usando sympy para evitar erros
s_0 = s.subs(t, 0)
s_1_5 = s.subs(t, 1.5)
delta_s = s_1_5 - s_0

print(f'Distância total percorrida nos primeiros 1,5 segundos: {delta_s.evalf()}')
```

# Questão 4 - Vale até 1 ponto

Encontre a área limitada por $f(x) = -x^2 + 4x + 1$, $g(x) = -x^3 + 7x^2 - 10x + 3$ para  $1\leq x \leq 2$. Nesta questão você pode usar álgebra, ou Python, considerando as limitações de usos de bibliotecas definidas para esta avaliação.  


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Definindo as funções
def f(x):
    return -x**2 + 4*x + 1

def g(x):
    return -x**3 + 7*x**2 - 10*x + 3

def h(x):
    return f(x) - g(x)

# Definindo o intervalo
a = 1
b = 2

# Calculando a integral
area, _ = quad(h, a, b)

# Plotando as funções
x = np.linspace(a, b, 400)
f_x = f(x)
g_x = g(x)

plt.plot(x, f_x, label=r'$f(x) = -x^2 + 4x + 1$')
plt.plot(x, g_x, label=r'$g(x) = -x^3 + 7x^2 - 10x + 3$')
plt.fill_between(x, f_x, g_x, where=(f_x > g_x), interpolate=True, color='lightgray', alpha=0.5)
plt.title('Área entre as curvas $f(x)$ e $g(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.grid(True)
plt.show()

print(f'Área calculada: {area}')
```
